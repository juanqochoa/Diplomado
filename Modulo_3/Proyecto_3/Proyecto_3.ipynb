```python
#https://data.mendeley.com/datasets/46htwnp833/2
#https://www.sciencedirect.com/science/article/abs/pii/S0925521420301629
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
```

```python
df["Set"].unique() #Vamos a ver que tipos de datos tenemos, hay de calibración y de evaluación
```

```python
valores_unicos_n = df.nunique()
columnas_sn_info = valores_unicos_n[valores_unicos_n == 1].index.tolist()
columnas_sn_info
```

```python
df_clean = df.drop(columnas_sn_info, axis=1)
```

```python
df_clean.describe()
```

```python
# Para seleccionar una variedad de Mango, se escoge "Caly"
df_clean["Cultivar"].unique()
```

```python
df_clean = df_clean[df_clean["Cultivar"] == "Caly"]
```

```python
df_clean.head()
```

```python
df_clean["Set"].unique() #Ahora vamos a usar solo los datos de validación "Val Ext"
```

```python
df_test = df_clean[df_clean["Set"] == "Val Ext"]
```

```python
df_cal = df_clean[df_clean["Set"] != "Val Ext"]
```

```python
df_cal["Set"].unique()
```

```python
df_cal.head()
```

```python
#Establecemos variable dependiente y las independientes
y_test = df_test["DM"]
y_cal = df_cal["DM"]

X_cal = df_cal.iloc[:,9:]
X_test = df_test.iloc[:,9:]
```

```python
#Importamos mismas librerías de PCA
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.cross_decomposition import PLSRegression
```

```python
PCR_Pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("pca", PCA(n_components=9)),
    ("regresion", LinearRegression())
])
MLR_Pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("regresion", LinearRegression())
])
```

```python
MLR_Pipeline.fit(X_cal,y_cal)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
# Métricas Regresión Lineal Multiple

pred_mlr = MLR_Pipeline.predict(X_test)
print(r2_score(y_test,pred_mlr))
print(np.sqrt(mean_squared_error(y_test,pred_mlr)))
```

```python
# Metricas PCR

pred_pcr = MLR_Pipeline.predict(X_test)
print(r2_score(y_test,pred_pcr))
print(np.sqrt(mean_squared_error(y_test,pred_pcr)))
```

```python
pca = PCR_Pipeline.named_steps["pca"]
explained_variance = pd.DataFrame(pca.explained_variance_ratio_)
cumsum = explained_variance.cumsum()
cumsum.plot()
```

```python
# Busqueda por regilia

from sklearn.model_selection import GridSearchCV, KFold
```

```python
# Busqueda por regilia

from sklearn.model_selection import GridSearchCV, KFold

pc_params = {
    "pca__n_components": np.arange(1,200)
}

cv = KFold(n_splits=10,shuffle=True) #n_splits es la división de subdatasets
pc_grid = GridSearchCV(PCR_Pipeline, pc_params, cv=cv, scoring="neg_mean_squared_error")
pc_grid.fit(X_cal,y_cal)
```

```python
pc_grid.best_params_["pca__n_components"]
```

```python
# Busqueda por regilla
pc_params = {
    "regresion__n_components": [32,64,100,150,200,250,280]
}

cv = KFold(n_splits=10,shuffle=True) #n_splits es la división de subdatasets
pc_grid = GridSearchCV(PCR_Pipeline, pc_params, cv=cv, scoring="neg_mean_squared_error")
pc_grid.fit(X_cal,y_cal)
pc_grid.best_params_["regresion__n_components"]
```

```python
PLS_Pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("regresion", PLSRegression())
])
PLS_Pipeline.fit(X_cal,y_cal)
```

```python
#Metricas PCR
pred_pcr = PLS_Pipeline.predict(X_test)
print(r2_score(y_test, pred_pcr))
print(np.sqrt(mean_squared_error(y_test,pred_pcr)))
```
