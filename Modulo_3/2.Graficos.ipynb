```python
#Grafica con dos lineas
experiencia = [1,3,5,7,9,13]
salario_data_scientists = [5000, 25000, 67000, 150000, 175000, 246354]
salario_ing_software = [1000, 6000, 15000, 37000, 99000, 145321]
plt.plot(experiencia, salario_data_scientists)
plt.plot(experiencia, salario_ing_software)
plt.title("Los Salarios") #Titulo
plt.xlabel("Experience") #Titulo eje x
plt.ylabel("Annual Income (USD)") #Titulo eje x=y
plt.legend(["Data Scientist", "Software Engineer"])
```

```python
plt.plot(experiencia, salario_data_scientists, color = "#990000", label = "Data Scientist", linestyle = "--", marker = "o") # Colores se pueden en Hex
plt.plot(experiencia, salario_ing_software, color = "#536de4", label = "Software Engineer", linestyle = ":", marker = "*")
plt.title("Los Salarios") #Titulo
plt.xlabel("Experience (Years)") #Titulo eje x
plt.ylabel("Annual Income (Trillion USD)") #Titulo eje x=y
plt.legend()
```

```python
#Grafica de barras
x = ["A", "B", "C", "D"]
y = [5,7,6,8]
plt.bar(x,y)
plt.title("Los Salarios") #Titulo
plt.xlabel("Experience (Years)") #Titulo eje x
plt.ylabel("Annual Income (Trillion USD)") #Titulo eje x=y
```

```python
#Grafica de tendencia
x = ["A", "B", "C", "D"]
y = [2,7,6,8]
tendencia = [2,7,6,8]
plt.bar(x,y)
plt.plot(x,tendencia, color = "m")
plt.title("Los Salarios") #Titulo
plt.xlabel("Experience (Years)") #Titulo eje x
plt.ylabel("Annual Income (Trillion USD)") #Titulo eje x=y
```

```python
import numpy as np

#Gráifico de dispersión
x = np.random.rand(100)
y = np.random.rand(100)
plt.scatter(x,y,color = "r", alpha = 0.6)
plt.title("Gráfico de dispersión")
```

```python
#Histograma
datos = np.random.randn(10000)
plt.hist(datos,bins=20)
plt.title("Histograma")
plt.xlabel("Experience (Years)") #Titulo eje x
plt.ylabel("Annual Income (Trillion USD)") #Titulo eje x=y
```

```python
import pandas as pd
data_range = pd.date_range(start = "2016-05-06", end = "2017-05-04", freq = "D")
precios = np.cumsum(np.random.randn(len(data_range))) + 100
```

```python
data = pd.DataFrame({"Date": data_range, "Precios": precios})
data.head()
```

```python
#Serie de tiempo
plt.plot(data["Date"],data["Precios"], label = "Precio")
```

```python
#Datos Grafico Lineas
x1 = np.linspace(0,10,100)
y1 = np.sin(x)
y12 = np.cos(x)
#Datos Grafico Barras
categorias = ["a", "b", "c", "d"]
valores = [5,7,3,8]
#Datos Grafico Dispersión
x2 = np.random.rand(100)
y2 = np.random.rand(100)
#Datos Histograma
datos = np.random.randn(999)
```

```python
#Cuadrantes
fig, axs = plt.subplots(2,2, figsize = (15,10))
#Datos Grafico Lineas
axs[0,0].plot(x1,y1, color = "#be2563", label = "Seno")
axs[0,0].plot(x1,y12, color = "#be2500", label = "Coseno")
axs[0,0].set_title("Gráfico de líneas")
axs[0,0].legend(loc = "upper center")
#Datos Grafico Barras
axs[0,1].bar(categorias, valores, color = "#2596BE")
axs[0,1].set_title("Gráfico de barras")
#Datos Grafico Dispersión
axs[1,0].scatter(x2,y2,color = "#2568be", alpha = 0.89)
axs[1,0].set_title("Gráfico de dispersión")
#Histograma
axs[1,1].hist(datos,bins=15, color = "#be2563", alpha = 0.75)
axs[1,1].set_title("Histograma")

fig.suptitle("Ejemplos de diferentes tipos de gráficos", fontsize = 20)
```
